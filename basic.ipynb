```python
from abc import ABC, abstractmethod
class Animal(ABC):
    @abstractmethod
    def Move(self):
        pass



class Bird(Animal):
    def Move(self):
        print("Bird")



class Cat(Animal):
    def Move(self):
        print("Cat")
```
